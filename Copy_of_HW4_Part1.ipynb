```python
#Please use this cell to create your your figure. Please use Year column to color your graph.
#Please use this cell to create your your figure. Please use Year column to color your graph.
import pandas as pd
import matplotlib.pyplot as plt

# Try reading the file without the 'error_bad_lines' argument
df_elections = pd.read_csv('/content/drive/MyDrive/DATA/2016elections.csv')

# Filter for states, considering both abbreviations and full names
states_to_include = ['AR', 'MI', 'CA', 'WI', 'Arkansas', 'Michigan', 'California', 'Wisconsin']
df_elections = df_elections[df_elections['state'].isin(states_to_include)]

# Select the relevant columns and set 'state' as index
df_elections = df_elections[['state', 'pct_clinton', 'pct_trump', 'pct_johnson', 'pct_other']].set_index('state')

# Now check if the DataFrame is still empty
if df_elections.empty:
    print("Warning: DataFrame is still empty after adjusting the query.")
else:
    print(df_elections.index)
    df_votes = df_elections[['pct_clinton', 'pct_trump', 'pct_johnson', 'pct_other']]

    ax = df_votes.plot(kind='bar', stacked=True, color=['blue', 'red', 'gold', 'grey'])
    ax.set(title='2016 Election Vote Percentages by State', xlabel='State', ylabel='Vote Percentage')
    plt.xticks(rotation=0)
    plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# create DataFrame
df = pd.DataFrame({'High Temp': [28, 30, 34, 38, 45, 42,
                                 38, 35, 32, 28, 25, 21],
                   'Low Temp': [22, 26, 30, 32, 41, 38,
                                32, 31, 28, 22, 15, 15],
                   'Avg Temp': [25, 28, 32, 35, 43, 40,
                                35, 33, 30, 25, 20, 18]},
                  index=['Jan', 'Feb', 'Mar', 'Apr', 'May',
                         'Jun', 'Jul', 'Aug', 'Sep', 'Oct',
                         'Nov', 'Dec'])


# create stacked bar chart for monthly temperatures
df.plot(kind='bar', stacked=True, color=['red', 'skyblue', 'green'])

# labels for x & y axis
plt.xlabel('Months')
plt.ylabel('Temp ranges in Degree Celsius')

# title of plot
plt.title('Monthly Temperatures in a year')
```
